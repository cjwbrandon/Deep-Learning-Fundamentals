# Hyperparameter Tuning

In this script, we will explore the different hyperparameters in a Neural Network and which you should pay more attention to. Given the vast number of hyperparameters and limited resources, we want to establish some sort of prioritisation for which HPs to be tuned first based on their potential impact to the NN. 

Depending on your application and setting, we might choose to tune a different number of HPs. As we tune more hyperparameters, the number of possible combinations grows exponentially.

For example, in practice, we often tune a hand full of the more important parameters. However, in a research, competition or precision-driven use case (medical related), we might tune more HPs just to squeeze those few points in accuracy.

A general guidelines for Hyperparamters prioritisation:

1. Highest
    - Learning rate
2. 2nd
    - Momentum term (if u use GD w/ momentum)
    - Number of hidden units
    - Mini-batch size
3. 3rd
    - Number of hidden layers
    - Learning rate decay
4. 4th - all the rest. You may prioritise differently depending on your application

In addition, I will also look at 2 methods to tune the models namely Grid Search and Random Sampling as well as Batch Normalisation that makes the NN more robust to the choice of HPs.

# Generate Data

```python
from sklearn.datasets import make_regression
```

```python
x, y = make_regression(n_samples=100000,
                       n_features=8,
                       n_informative=6,
                       n_targets=1,
                       noise=100,
                       bias=200,
                       random_state=2020
                      )
```

```python
x = x*100 + 100
y += 500
```

```python
x = torch.from_numpy(x).float()
y = torch.from_numpy(y).float()
```

```python
# split to train & validation set
split = int(0.9 * 100000)
x_train = x[:split]
x_val = x[split:]
y_train = y[:split]
y_val = y[split:]
```

# Learning Rate

Learning rate controls the size of our steps each time we update our weights. A large learning rate could potentially speed up learning but reduces precision in the convergence. A small learning rate could result in inefficient training. Finding a balance is essential as the learning rate can essentially result in a successful or unsuccessful Network.

```python
def mini_batch_training(nn_model, batch_size, epochs):
    # Training the Network
    loss = []
    val_loss = []

    for e in range(epochs):

        permutation = torch.randperm(x_train.size()[0]) # random permutation to group batches differently

        for i in range(0, x_train.size()[0], batch_size):
            optimiser.zero_grad() # zero out any pre-computed gradients

            # assign batch
            indices = permutation[i:i+batch_size]
            batch_x, batch_y = x_train[indices], y_train[indices]
            batch_x = batch_x.cuda()
            batch_y = batch_y.cuda()

            output = nn_model(batch_x) # forward prop
            run_loss = criterion(output, batch_y) # compute loss
            loss.append(run_loss.item()/batch_x.size()[0]) # store loss
            run_loss.backward() # backprop - calculate gradients
            optimiser.step() # update weights

        permutation = torch.randperm(x_val.size()[0]) # random permutation to group batches differently

        for i in range(0, x_val.size()[0], batch_size):

            # assign batch
            indices = permutation[i:i+batch_size]
            batch_x, batch_y = x_train[indices], y_train[indices]
            batch_x = batch_x.cuda()
            batch_y = batch_y.cuda()

            output = nn_model(batch_x)
            run_loss = criterion(output, batch_y)
            val_loss.append(run_loss.item()/batch_x.size()[0])
            
    # plot training and validation loss over training
    plt.figure(figsize=[20,10])
    plt.plot(range(len(loss)), loss)
    plt.show()
    plt.figure(figsize=[20,10])
    plt.plot(range(len(val_loss)), val_loss)
    plt.show()
```

```python
learning_rates = [1e-1, 1e-2, 1e-3, 1e-4]

for lr in learning_rates:
    print('---Current Learning Rate: {}---'.format(lr))
    # Network architecture
    nn_model = nn.Sequential(
        nn.Linear(8, 16),
        nn.ReLU(),
        nn.Linear(16, 1)
    )

    # move model & dataset to gpu
    nn_model = nn_model.cuda()

    # loss & optimiser
    criterion = nn.MSELoss()
    optimiser = optim.Adam(nn_model.parameters(), lr=lr)

    # Batch GD (full dataset)
    mini_batch_training(nn_model, batch_size=1028, epochs=1000)
    print()
```

# Momentum

```python
moms = [0.9, 0.99, 0.999]

for mom in moms:
    print('---Current Momentum HP: {}---'.format(mom))
    # Network architecture
    nn_model = nn.Sequential(
        nn.Linear(8, 16),
        nn.ReLU(),
        nn.Linear(16, 1)
    )

    # move model & dataset to gpu
    nn_model = nn_model.cuda()

    # loss & optimiser
    criterion = nn.MSELoss()
    optimiser = optim.SGD(nn_model.parameters(), lr=1e-2, momentum=mom)

    # Batch GD (full dataset)
    mini_batch_training(nn_model, batch_size=1028, epochs=100)
    print()
```
