# Standardisations & Normalisations

In this script, I will be exploring the different types of Normlisation and Standardisation techniques implemented in sklearn. Normalisations or standardisations are a common pre-processing step when training a Network as it can potentially make training much easier (faster) using the Gradient Descent algorithm. The main idea behind normalisations is getting all your input variables to be of the same scale, preferably small. On top of that, these techniques also helps us deal with outliers of our data.

1. Same scale
    - Often, we normalise the input data to 0 mean and a standard deviation of 1 (standard normal).
    - However, the main idea should be that we want all the variables to be of the same scale. This is to create an error space that is easier for your gradient descent to converge.
2. Small magnitudes
    - The reason for small magnitudes is largely inline with our activation functions where smaller values usually lands on the linear portion of the function. This would make training faster as gradients are larger.
    
    
We will be exploring the different standardisations and normalisations methods based on sklearn library:
1. Standard Scaler
    - Standard rescaling to mean 0 and s.d. 1
    - z = (x-u) / s
2. Min-Max Scaler
    - Transform features by scaling each feature to 0 to 1 (default)
    - z = (x - x{min}) / (x{max} - x{min})
3. Max-Absoulute Scaler
    - Scale each feature by its maximum absolute values. Does not shift the mean and deos not destroy sparsity.
    - Range from -1 to 1
    - z = x / max(abs(x))
4. Robust Scaler
    - Scale features using statistics that are robust to outliers. Removes median and scales the data according to quantile range.
5. Quantile Transformer
    - Transform features using quantiles information. (Non-linear transformation)
    - Normal - Features are mapped to a normal distribution
    - Uniform - Features are mapped to a uniform distribution
7. Power Transformer Yeo-Johnson
    - Transform features to be more Gaussian-like. 0 mean, 1 s.d.
    - Yeo-Johnson - supports both positive and negative data
8. Normaliser

Reference: https://towardsdatascience.com/normalization-vs-standardization-quantitative-analysis-a91e8a79cebf

# Generate data

```python
from sklearn.datasets import make_classification
from sklearn.metrics import classification_report
```

```python
# create a classification problem using sklearn dataset
x, y = make_classification(n_samples=10000,
                           n_features=4,
                           n_informative=4,
                           n_redundant=0,
                           n_classes=2,
                           flip_y=0.2,
                           random_state=2020
                          )
```

```python
for i in range(4):
    plt.figure(figsize=[20,10])
    plt.hist(x[:, i], bins = int(10000/100), color='purple')
    plt.show()
```

Lets adjust the mean and variance of our features to observe how they can affect learning.

```python
x[:,0] = x[:,0] + 100 # increase mean
x[:,1] = x[:,1] * 4 # increase s.d.
x[:,2] = x[:,2] - 50 # decrease mean
x[:,3] = (x[:,3] * 3) + 100 # increase mean and s.d
```

```python
# transform to tensors
x = torch.from_numpy(x).float()
y = torch.from_numpy(y).long()
```

```python
# split to train & validation set
split = int(0.9 * 10000)
x_train = x[:split]
x_val = x[split:]
y_train = y[:split]
y_val = y[split:]
```

# No Normalising or standardising

```python
# Network architecture
nn_model = nn.Sequential(
    nn.Linear(4, 64),
    nn.Tanh(),
    nn.Linear(64,128),
    nn.Tanh(),
    nn.Linear(128, 32),
    nn.Tanh(),
    nn.Linear(32, 2)
)

# move model & dataset to gpu
nn_model = nn_model.cuda()
x_train = x_train.cuda()
y_train = y_train.cuda()
x_val = x_val.cuda()
y_val = y_val.cuda()

# loss & optimiser
criterion = nn.CrossEntropyLoss()
optimiser = optim.SGD(nn_model.parameters(), lr =0.001)
```

```python
%%time
# Training the Network
epochs = 15000
loss = []
val_loss = []

for e in range(epochs):
    optimiser.zero_grad() # zero out any pre-computed gradients

    output = nn_model(x_train) # forward prop
    run_loss = criterion(output, y_train) # compute loss
    loss.append(run_loss.item()/x_train.size()[0]) # store loss
    
    run_loss.backward() # backprop - calculate gradients
    optimiser.step() # update weights
    
    output = nn_model(x_val)
    run_loss = criterion(output, y_val)
    val_loss.append(run_loss.item()/x_val.size()[0])
```

```python
# plot training and validation loss over training
plt.figure(figsize=[20,10])
plt.plot(range(len(loss)), loss)
plt.show()
plt.figure(figsize=[20,10])
plt.plot(range(len(val_loss)), val_loss)
plt.show()
```

Observe how it took approximately >10,000 epochs for the loss function to converge and even so, at the end, we see the model faces difficulty converging due to the learning rate causing the zig-zag formation.

```python
# move model & dataset to cpu
nn_model = nn_model.cpu()
x_train = x_train.cpu()
y_train = y_train.cpu()
x_val = x_val.cpu()
y_val = y_val.cpu()

# Training accuracy
output = nn_model(x_train)
_, predictions = torch.max(output, axis=1)
print(classification_report(y_train, predictions))

# Val accuracy
output = nn_model(x_val)
_, predictions = torch.max(output, axis=1)
print(classification_report(y_val, predictions))
```

# throw all into a function

```python
def full_network(x):
    # transform to tensors
    x = torch.from_numpy(x).float()

    # split to train & validation set
    split = int(0.9 * 10000)
    x_train = x[:split]
    x_val = x[split:]
    y_train = y[:split]
    y_val = y[split:]

    # Network architecture
    nn_model = nn.Sequential(
        nn.Linear(4, 64),
        nn.Tanh(),
        nn.Linear(64,128),
        nn.Tanh(),
        nn.Linear(128, 32),
        nn.Tanh(),
        nn.Linear(32, 2)
    )

    # move model & dataset to gpu
    nn_model = nn_model.cuda()
    x_train = x_train.cuda()
    y_train = y_train.cuda()
    x_val = x_val.cuda()
    y_val = y_val.cuda()

    # loss & optimiser
    criterion = nn.CrossEntropyLoss()
    optimiser = optim.SGD(nn_model.parameters(), lr =0.001)

    # Training the Network
    epochs = 15000
    loss = []
    val_loss = []

    for e in range(epochs):
        optimiser.zero_grad() # zero out any pre-computed gradients

        output = nn_model(x_train) # forward prop
        run_loss = criterion(output, y_train) # compute loss
        loss.append(run_loss.item()/x_train.size()[0]) # store loss

        run_loss.backward() # backprop - calculate gradients
        optimiser.step() # update weights

        output = nn_model(x_val)
        run_loss = criterion(output, y_val)
        val_loss.append(run_loss.item()/x_val.size()[0])

    # plot training and validation loss over training
    plt.figure(figsize=[20,10])
    plt.plot(range(len(loss)), loss)
    plt.show()
    plt.figure(figsize=[20,10])
    plt.plot(range(len(val_loss)), val_loss)
    plt.show()

    # move model & dataset to cpu
    nn_model = nn_model.cpu()
    x_train = x_train.cpu()
    y_train = y_train.cpu()
    x_val = x_val.cpu()
    y_val = y_val.cpu()

    # Training accuracy
    output = nn_model(x_train)
    _, predictions = torch.max(output, axis=1)
    print(classification_report(y_train, predictions))

    # Val accuracy
    output = nn_model(x_val)
    _, predictions = torch.max(output, axis=1)
    print(classification_report(y_val, predictions))
```

# StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
# scale variables
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# New distribution
for i in range(4):
    plt.figure(figsize=[20,10])
    plt.hist(x_scaled[:, i], bins = int(10000/100), color='purple')
    plt.show()
```

Observe how they are not centered at mean 0 and s.d. 1. However, we note that some of the features are skewed towards the left/right.

Lets experiment how this helps with our learning

```python
%%time
full_network(x_scaled)
```

With the standardisation, we observe a much smoother curve where the cost function continues to converge to a lower magnitude as compared to before. We can consider increasing the learning rate to make learning faster which is not possible in the previous case as it would accelerate and amplify the zig-zag formations seen.

In addition, as it is able to learn faster, we observe a slightly better accuracy with the same hyperparameters.

We observe that the learning was linear and slower as compared to all the previous models. Only near the end, we observe that the function starts to converge at a much higher magnitude. One difference between this transformation and the rest is that it does not alter the center or the sparsity of the dataset. This might have impacted the benefit of the standardisation. However, it could also be that this particular dataset does not go well with this particular standardisation. More have to be explored to be conclusive.
